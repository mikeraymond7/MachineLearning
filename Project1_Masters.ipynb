**Mine From StockAnalysis**

```python
def passes(ticker):
  return ticker.isalpha() and len(ticker) < 5
```

```python
# test mining
import time
import random
# note: purposely left out try, catch blocks so the program stopped when 429 was received

all_etfs = pd.DataFrame(stocks.find({})).set_index('ETF').to_dict()
finished_ticks = list(all_etfs['_id'].keys())
print(finished_ticks)
x = 374
for i, tick in enumerate(tickers[x:]):
  if not(tick in finished_ticks):
    '''if i%10==0 and i != 0:
      time.sleep(60)
    time.sleep(6)'''
    url = f'https://stockanalysis.com/etf/{tick}/holdings/'
    etf_data = {'ETF': [], 'Holding Ticker': [], 'Weight': [], 'NumShares': []}
    #headers={'User-Agent': generate_user_agent()}
    headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36'}
    req = Request(url=url,headers=headers)
    resp = urlopen(req)
    #print(f'Response for {tick}, {resp.status}')

    html = BeautifulSoup(resp, features="lxml")
    equity = html.text.find('Equity')
    if equity > 0:
      tr = html.find_all('tr')[1:]
      # only take if hrefs
      for row in tr:
        td = row.find_all('td')
        data = [data.text for data in td]
        #print(data)
        if passes(td[1].text):
          etf_data['ETF'].append(tick)
          etf_data['Holding Ticker'].append(data[1])
          etf_data['Weight'].append(data[3][:-1])
          etf_data['NumShares'].append(data[4])
        #else:
          #print(f'{data[1]} does not pass for {tick}')
      if len(etf_data['ETF']) > 0:
        print(pd.DataFrame(etf_data))
        stocks.insert_many(pd.DataFrame(etf_data).to_dict('records'))
      else:
        time.sleep(2)
  else:
    print(f'{tick} already in database')
  print(f'{tick}, {i+x}')
```

```python
# returns set of all ETF tickers as a list
all_funds = pd.DataFrame(stocks.find({}))
def good_tickers():
  return list(pd.DataFrame(stocks.find({})).set_index('ETF').to_dict()['_id'].keys())
good_tics = good_tickers()
good_tics.sort()

# returns set of holding tickers for a given ETF
def get_set_holdings(etf):
  try:
    return set(all_funds['Holding Ticker'].loc[all_funds['ETF']== etf])
  except:
    print(f'Failed to get holdings for {etf}')
    return set([])

# Standard jaccard function takes 2 sets
def jaccard(set1,set2):
    return len(set1.intersection(set2))/len(set1.union(set2))

# Checks jaccard similarity between all ETFs in db
def jaccard_all_etfs():
  for i in range(len(good_tics)):
    for j in range(i+1,len(good_tics)):
      tic1 = good_tics[i]
      tic2 = good_tics[j]
      holdings1 = get_set_holdings(tic1)
      holdings2 = get_set_holdings(tic2)
      jac = jaccard(holdings1, holdings2)
      print(f'{tic1} --> {tic2} ====> {jac}')

# Calculates all Jaccard similarities for one ETF against all others and prints outputs
def allJaccardETF(etf):
  print('|\tETF1\t|\tETF2\t|\tSimilarity\t|\n --------------------------------------------------------')
  etfSet1 = get_set_holdings(etf)
  for i in good_tics:
    if i != etf:
      etfSet2 = get_set_holdings(i)
      print(f'|\t{etf}\t|\t{i}\t|\t{jaccard(etfSet1, etfSet2)}')
```

```python
all_expense_ratios = pd.DataFrame(expenseRatios.find({}))

def recommend_etf(etf, topn):
  jac = {'ETF':[],'Holding Tickers':[], 'Jaccard': []}
  etfSet1 = get_set_holdings(etf)
  for i in good_tics:
    if i != etf:
      etfSet2 = get_set_holdings(i)
      jac['ETF'].append(etf)
      jac['Holding Tickers'].append(i)
      jac['Jaccard'].append(jaccard(etfSet1, etfSet2))
  jac_df = pd.DataFrame.from_dict(jac).sort_values('Jaccard', ascending = False)
  exp_rt = list(all_expense_ratios['Expense Ratio'].loc[all_expense_ratios['ETF'] == etf].to_dict().values())[0]
  expense_recommend = {'ETF': [f'Original: {etf}'], 'Jaccard': [1], 'Expense Ratio': [exp_rt]}
  rec_dict = jac_df[:topn].to_dict()
  for key,etf in rec_dict['Holding Tickers'].items():
    exp_rt = list(all_expense_ratios['Expense Ratio'].loc[all_expense_ratios['ETF'] == etf].to_dict().values())
    expense_recommend['ETF'].append(etf)
    expense_recommend['Jaccard'].append(rec_dict['Jaccard'][key])
    if len(exp_rt) > 0:
      expense_recommend['Expense Ratio'].append(exp_rt[0])
    else:
      expense_recommend['Expense Ratio'].append('n/a')
  return pd.DataFrame(expense_recommend)

```
